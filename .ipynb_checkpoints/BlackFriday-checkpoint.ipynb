Let's find out if there exist any missing values in our data.

```python
total=data.isnull().sum().sort_values(ascending=False)
percent=(data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
missing= pd.concat([total, percent], axis=1, join='outer',keys=['Total missing count', 'Percentage '])
missing
```

As we can see, about 69% information is missing in Product_Category_3 and 31% in Product_Category_2.

Let's drop these rows.

```python
data.dropna(0, inplace=True)
```

```python
data.isnull().sum()
```

Let's visualize each feature one by one.

##### Let's view who is more likely to spend more on Black Friday i.e Man/Woman

```python
sns.countplot(data['Gender'])
```

```python
plt.figure(figsize=[5,5])
plt.pie(data['Gender'].value_counts(),labels=['Female','Male'],explode=(0.1,0),
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)
plt.tight_layout()
plt.legend()
plt.show()
```

Above plots show that men spend more on Balck Friday than women.

###### Let's view which age group people are more likely to spend more on Black Friday

```python
plt.figure(figsize=[9,6])
sns.countplot(data['Age'])
```

```python
plt.figure(figsize=[9,6])
sns.countplot(data['Age'], hue=data['Gender'])
```

It clearly shows that men in every age group are spending more money on Black Friday sales than women. Where are people who say women spend more money on sales? See this! 

```python
plt.figure(figsize=[9,6])
data.groupby('Age')['Purchase'].sum().sort_values().plot('bar')
```

Above plots show that people of age group 26-35 are spending more on Black Friday than all other age groups. It aslo shows that kids are less likely to spend their pocket money on Black Friday sales :p

##### Let's find out Marital status of people who spend more on Black Friday

```python
data['Marital_Status']=data['Marital_Status'].replace(1, 'married')
data['Marital_Status']=data['Marital_Status'].replace(0, 'ummarried')
plt.figure(figsize=[9,6])
sns.countplot(data['Marital_Status'], hue=data['Gender'])
```

```python
plt.figure(figsize=[5,5])
plt.pie(data['Marital_Status'].value_counts(),labels=['unmarried','married', ],explode=(0.1,0),
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)
plt.tight_layout()
plt.legend()
plt.show()
```

Seems like unmarried people have alot to spend on sales than married ones. Don't get married, save money and spend on shopping! :D

##### Let's visualize location of people wo spend more on Black Friday

```python
plt.figure(figsize=[9,6])
sns.countplot(data['City_Category'])
```

```python
plt.figure(figsize=[5,5])
plt.pie(data['City_Category'].value_counts(),labels=['A', 'B', 'C'],
        explode=(0.1,0,0),
        shadow=True,
        startangle=90,
       autopct='%1.1f%%')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[5,5])
plt.pie(data.groupby('City_Category')['Purchase'].sum(),labels=data['City_Category'].unique(),
        explode=(0.1,0,0),
        shadow=True,
        startangle=90,
       autopct='%1.1f%%')
plt.tight_layout()
plt.legend()
plt.show()
```

It shows that people who located in C are spending more on Black Friday.

##### Let's visualize old and new customer's ratio

```python
plt.figure(figsize=[9,6])
sns.countplot(data['Stay_In_Current_City_Years'])
```

```python
plt.figure(figsize=[5,5])
plt.pie(data.groupby('Stay_In_Current_City_Years')['Purchase'].sum(),labels=data['Stay_In_Current_City_Years'].unique(),
        explode=(0.1,0.1,0,0,0),
        shadow=True,
        startangle=90,
       autopct='%1.1f%%')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[9,6])

data.groupby('Stay_In_Current_City_Years')['Purchase'].sum().sort_values().plot('bar')
```

Above plots show that people who are newly shifted to this city spend more on Black Friday sales. It's a good choice to buy all that home related stuff at cheaper rates on Black Friday! It also shows that people who are staying as guest in this city are less likey to spend money on sales like Black Friday.
